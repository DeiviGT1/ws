```python
#create a code that using selenium opens a browser and goes to starplus.com

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time
import os
from dotenv import load_dotenv

load_dotenv()

user_name = os.getenv("PRIME_VIDEO_USERNAME")
password = os.getenv("PRIME_VIDEO_PASSWORD")

driver = webdriver.Chrome()
driver.get("https://www.primevideo.com/")


#click on the login button
```

```python
path = './/a[@class="dv-copy-button"]'
login_button = driver.find_element(By.XPATH, path)
login_button.click()
```

```python
path = './/input[@type="email"]'
email_field = driver.find_element(By.XPATH, path)
email_field.send_keys(user_name)
```

```python
path = './/input[@id="continue"]'
continue_button = driver.find_element(By.XPATH, path)
continue_button.click()
```

```python
path = './/input[@type="password"]'
password_field = driver.find_element(By.XPATH, path)
password_field.send_keys(password)
```

```python
path = './/input[@id="signInSubmit"]'
sign_in_button = driver.find_element(By.XPATH, path)
sign_in_button.click()
```

```python
path = './/input[@aria-label="Buscar en Prime Video"]'
search_field = driver.find_element(By.XPATH, path)
search_field.send_keys("two and a half men")
#click enter
```

```python
path = './/div[@class="dv-node-dp-seasons"]/div/div/ul/li/a'
seasons = driver.find_elements(By.XPATH, path)
print(len(seasons))
for season in seasons:
    #get a and its href
    print(season.get_attribute("href"))
    driver.get(season.get_attribute("href"))
    time.sleep(3)
```

```python
path = './/div[@class="dCocJw"]'
episodes = driver.find_elements(By.XPATH, path)
eps = {}
current_episode = 1
for episode in episodes:
    link = episode.find_element(By.XPATH, './div/a').get_attribute("href")
    eps[f"{current_episode}"] = {"link": "" + link, 
                                "chapter_name": episode.text}
    
    current_episode += 1
print(eps)
```

```python
path = './/div[@class="klzoqL"]/div/ul/li/a'
seasons = driver.find_elements(By.XPATH, path)
s = []
for season in seasons:
    s.append(season.get_attribute("href"))


temp = {}
current_season = 1
for season in s:
    driver.get(season)
    print(season)
    time.sleep(3)
    path = './/div[@class="dCocJw"]'
    episodes = driver.find_elements(By.XPATH, path)
    eps = {}
    current_episode = 1
    for episode in episodes:
        try:
            link = episode.find_element(By.XPATH, './div/a').get_attribute("href")
            eps[f"episode_{current_episode}"] = {"link": "" + link, 
                                    "chapter_name": episode.text}
        except:
            eps[f"episode_{current_episode}"] = {"link": "", 
                                    "chapter_name": ""}
        current_episode += 1
    
    temp[f"season_{current_season}"] = eps
    current_season += 1
```
